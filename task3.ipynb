```python
import pandas as pd

data = {
    'student_name': ['Ivan', 'Evgeniy', 'Aleksey', 'Eva', 'Grigoriy'],
    'attendance': [95, 80, 75, 90, 85],
    'mark': [88, 75, 60, 92, 80]
}

df = pd.DataFrame(data)
print(df)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['attendance'], df['mark'], color='blue')

plt.title('Correlation between Attendance and Marks')
plt.xlabel('Attendance (%)')
plt.ylabel('Marks (out of 100)')

plt.grid()

plt.show()
```
